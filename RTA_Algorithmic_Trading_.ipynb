```python
#Extract Time of Daily Highs and Lows

AAPL_market.groupby('tradedate').agg({'low': min, 'high': max})
```

```python
AAPL_market.loc[AAPL_market.groupby('tradedate')['low'].idxmin()]
```

```python
AAPL_market.loc[AAPL_market.groupby('tradedate')['high'].idxmax()]
```

```python
#Adding moving average to the dataframe

AAPL_market['1hr-interval'] = AAPL_market['close'].rolling(4).mean()
AAPL_market['3hrs-interval'] = AAPL_market['close'].rolling(12).mean()

AAPL_market
```

```python
#Adding "signal" column

import numpy as np

AAPL_market['signal'] = np.where(AAPL_market['1hr-interval'] > AAPL_market['3hrs-interval'], 1, 0)
AAPL_market['signal'] = np.where(AAPL_market['1hr-interval'] < AAPL_market['3hrs-interval'], -1, AAPL_market['signal'])
AAPL_market.dropna(inplace=True)
AAPL_market.head()
```

```python
#Calculate instantaneus/system returns

AAPL_market['return'] = np.log(AAPL_market['close']).diff()
AAPL_market['system_return'] = AAPL_market['signal'] * AAPL_market['return']
AAPL_market['entry'] = AAPL_market.signal.diff()

AAPL_market.head()
```

```python
#Plotting trades on time series

plt.rcParams['figure.figsize'] = 12,6
plt.grid(True, alpha = .3)
plt.plot(AAPL_market.iloc[-252:]['close'], label = 'price at close')
plt.plot(AAPL_market.iloc[-252:]['1hr-interval'], label = '1hr-interval')
plt.plot(AAPL_market.iloc[-252:]['3hrs-interval'], label = '3hrs-interval')
plt.plot(AAPL_market[-252:].loc[AAPL_market.entry==2].index, AAPL_market[-252:]['3hrs-interval'][AAPL_market.entry==2], '^',
         color = 'g', markersize = 12)
plt.plot(AAPL_market[-252:].loc[AAPL_market.entry==-2].index, AAPL_market[-252:]['3hrs-interval'][AAPL_market.entry==-2], 'v',
         color = 'r', markersize = 12)
plt.legend(loc=1)

plt.title('Algorithmic trading of APPL stocks')
```

```python
#Plotting buy/hold returns vs system/algorithmic returns

plt.plot(np.exp(AAPL_market['return']).cumprod(), label = 'Buy/Hold')
plt.plot(np.exp(AAPL_market['system_return']).cumprod(), label = 'System')
plt.legend(loc=2)
plt.grid(True, alpha = .3)

plt.title('Buy/hold returns vs system returns of APPL stocks')
```
